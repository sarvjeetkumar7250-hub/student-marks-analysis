```python

import pandas as pd

data = {
    "Name": ["Aman", "Riya", "Rahul", "Sneha", "Arjun"],
    "Maths": [85, 90, 78, 92, 88],
    "Science": [80, 85, 75, 95, 82],
    "English": [78, 88, 70, 90, 85],
    "Attendance": [90, 95, 85, 98, 92]
}

df = pd.DataFrame(data)
df
```

```python
df["Average"] = df[["Maths", "Science", "English"]].mean(axis=1)
df
```

```python
top_student = df.loc[df["Average"].idxmax()]
print("Top Student:")
print(top_student)
```

```python

import matplotlib.pyplot as plt

plt.bar(df["Name"], df["Average"])
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Student Performance")
plt.show()
```

# Conclusion
- Sneha is the top-performing student
- Students with higher attendance scored better
- Performance depends on multiple subjects
