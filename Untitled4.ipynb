```python
import pandas as pd

data = {
    "Team": ["MI", "CSK", "RCB", "MI", "CSK", "RCB", "MI"],
    "Runs": [180, 200, 150, 170, 210, 160, 190],
    "Wickets": [5, 4, 6, 7, 3, 8, 5],
    "Result": ["Win", "Win", "Loss", "Win", "Win", "Loss", "Win"]
}

df = pd.DataFrame(data)
df
```

```python

# Total matches played by each team
matches = df["Team"].value_counts()
print(matches)

# Average runs by team
avg_runs = df.groupby("Team")["Runs"].mean()
print(avg_runs)
```

```python

import matplotlib.pyplot as plt

# Matches count graph
matches.plot(kind="bar")
plt.title("Matches Played by Teams")
plt.xlabel("Team")
plt.ylabel("Matches")
plt.show()
```

```python

avg_runs.plot(kind="bar")
plt.title("Average Runs by Team")
plt.xlabel("Team")
plt.ylabel("Runs")
plt.show()
```

- MI has played the most matches

- Teams with higher runs tend to win more matches
